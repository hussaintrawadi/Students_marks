```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.tail(15)
```

```python
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")

plt.show()
```

From the above chart we have analysed that the number of female students is more as compared to male students

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb
```

```python
sns.heatmap(gb, annot = True)
plt.title("Parent Education")
plt.show()
```

So from the above grap we can see that the Education of parents have a great impact on student's avrage score.
The score of the studet whose parents have a higher degree is greater as compared to others

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Parent's Marital Status")
plt.show()
```

From the above chart we can conclude that there is no impact of Parent's marital status on student score

```python
print(df["EthnicGroup"].unique())
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()
```

```python
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
l = ["Group A","Group B","Group C","Group D","Group E"]
color = ("#F99417","#B31312","#EA906C","#FF6C22","#2B3499")
plt.pie(mlist, labels = l, autopct = "%1.2f%%", colors=color)
plt.title("Distribution of Groups")
plt.show()
```

In conclusion we can say that the degree of students affect on the scores of students in a good way and their martial status does not affect the score of their children. 

Also we can conclude that students find Maths subject more difficult as compared to other subjects like Reading & Writing.

Lastly we can conclude that as per changing times now people have also started educating thier girl child as the numbers of Female students is comparatively more then male students
